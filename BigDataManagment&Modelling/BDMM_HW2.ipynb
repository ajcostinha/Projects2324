**CREATING TRANSACTIONS:**

We have decided to create a collection with transactions per listing. Once each document has referenced the listing_id, we don't need to have this information referenced on the listing collection side.

We embed the host information. Every field that starts with 'host' will be embeded onto an array called 'host_information'. Since we couldn't use regex for unseting fields, we decided to do a loop instead. This was done in the case of the impossibility of manually select the fields. If so, it would be much preferable, and therefore this code can be found below as well.

Our pipeline took into account the `regex` operator, found in this documentation: https://www.mongodb.com/docs/manual/reference/operator/query/regex/

```python
#Attribute Pattern in listings: host...
cursor = db.listings.find()
field_set = set()
for doc in cursor:
    for key in doc.keys():
        if key.startswith('host'):
            field_set.add(key)

unset_fields = list(field_set)

pipeline = [
    {
        '$set': {
            'host_information': {
                '$arrayToObject': {
                    '$filter': {
                        'input': {
                            '$objectToArray': '$$ROOT'
                        },
                        'as': 'field',
                        'cond': {
                            '$regexMatch': {
                                'input': '$$field.k',
                                'regex': '^host',
                                'options': 'i'  
                            }
                        }
                    }
                }
            }
        }
    },
    {
        '$unset': unset_fields  
    },
    {
        '$merge': {
            'into': 'listings',
            'whenMatched': 'replace'
        }
    }
]

db.listings.aggregate(pipeline)
```

```python
query_1 = {'$match': {'address.market': {'$ne': ''}}}

query_2 = {'$group':{
                '_id': {'city': '$address.market', 'accommodates': '$accommodates'}, #need to be both ids, otherwise it wont group the number of people
                'max_dif': {'$max': {'$abs': {'$subtract': ['$price', '$security_deposit']}}},
                'min_dif': {'$min': {'$abs': {'$subtract': ['$price', '$security_deposit']}}}
    }}

query_3 = {'$match': {
                'max_dif': {'$ne': None},
                'min_dif': {'$ne': None}
    }}

query_4 = {'$group': {
                '_id': '$_id.city',
                'differences': {
                    '$push': {
                        'accommodates': '$_id.accommodates',
                        'max_dif': '$max_dif',
                        'min_dif': '$min_dif'}}
    }}

pipeline=[query_1, query_2, query_3, query_4]

result = db.listings.aggregate(pipeline)
pprint(list(result))
```

- With `project` we return the dictionary with this new organization and after we user `sort` so we can properly see the distribution per property type.

```python
#Average stay per city
query_1 = {"$group":{
        "_id":{"key1":"$property_type", "key2":"$address.market"},
        "average_stay":{"$avg":"$maximum_nights"}
    }}

query_2 = {"$project": {
        "_id": 0,
        "property_type":"$_id.key1",
        "city":"$_id.key2",
        "average_stay":1,
    }}

query_3 = {"$sort": {
        "property_type":1,
        "average_stay":-1
    }}

pipeline = [query_1, query_2, query_3];

result = list(db.listings.aggregate(pipeline))
pprint(result)
```

- We notice that in some cities the average stay is **greater that 365** which we conclude those cases  may be refering to an eventual rent. We consider that we are analysing per year, so we recode it, but this time only with listings that have ``maximum_nights`` less then or equal to 365.

```python
query_0 = {"$match": {
        "maximum_nights": {"$lte": 365}
    }}

query_1 = {"$group":{
        "_id":{"key1":"$property_type", "key2":"$address.market"},
        "average_stay":{"$avg":"$maximum_nights"}
    }}

query_2 = {"$project": {
        "_id": 0,
        "property_type":"$_id.key1",
        "city":"$_id.key2",
        "average_stay":1,
    }}

query_3 = {"$sort": {
        "property_type":1,
        "average_stay":-1
    }}

pipeline = [query_0, query_1, query_2, query_3];

result = list(db.listings.aggregate(pipeline))
pprint(result)
```

**Highest and lowest average stay per property type**

- Using the **previous code** we look for the `first` and `last` observation per property type, considering the first the highest and the last the lowest since we sorted the values by property_type ascending and average_stay descending.

```python
query_4 = {"$group": {
    "_id": "$property_type",
    "max_city": {"$first": "$city"},
    "max_avg_stay": {"$first": "$average_stay"},
    "min_city": {"$last": "$city"},
    "min_avg_stay": {"$last": "$average_stay"}
}}

query_5 = {"$project": {
    "_id": 0,
    "property_type": "$_id",
    "max_city": 1,
    "max_avg_stay": 1,
    "min_city": 1,
    "min_avg_stay": 1
}}

query_6 = {"$sort": {
        "property_type":1,
    }}

pipeline = [query_1, query_2, query_3, query_4, query_5, query_6]
result = list(db.listings.aggregate(pipeline))
pprint(result)
```

### Question 12

We wish to have a summary webpage that displays information about our top destinations. This webpage should display for each of the top 10 cities some basic information about our operations in the area (number of properties by type for example, average price by type) but you can choose the metrics. For each of the top 10 cities it should also provide some basic information about the top 3 properties in each city (price, number of review, whatever you think useful) to show an example of the properties available in the area. We would like to keep this webpage up to date as information changes.

Analysing the code below stage by stage:

* 1st `match`: We previously knew there was a city '', so in order to not count it, we had to match all documents that weren't equal to ''.

* 2nd `group`:  it's ID-ing by city and type of property, calculating the average price, total of properties and reviews.

* 3rd `group`: it's grouping by city, placing property_types in an array with the total of properties the average price. And grouping with the total number of reviews and the total properties by city.

* 4th `sort`: sorting by the descending order of total reviews by city, the previous output.

* 5th `limit`: limiting the output by ten, in order of getting the top ten cities. In here, we decided to use total reviews as the metric of defining what would make the top ten - popularity.

* 6th `lookup`: to get the top three properties with the most reviews, this stage is joining the top ten cities with the listings collection. The double $ is used when refering to a variable defined in previous stages of the aggregation pipeline.

* 7th `project`: this stage simply projects the output as we define it.

```python
query_1 = {'$match': {'address.market': {'$ne': ''}}} 

query_2 = {'$group': {
            '_id': {'city': '$address.market', 'type': '$property_type'}, 
            'total_properties': {'$sum': 1},
            'total_reviews': {'$sum': {'$size': '$reviews'}}, 
            'average_price': {'$avg': '$price'}}
          }
    
query_3 = {'$group': {
            '_id': '$_id.city', 
            'property_types': {
                '$push': {
                    'type': '$_id.type', 
                    'count': '$total_properties',
                    'average_price':{'$round': [{'$avg': '$average_price'}, 2]}}},
            'total_reviews': {'$sum': '$total_reviews'},
            'total_properties': {'$sum': '$total_properties'}}
    }
    
query_4 = {'$sort': {'total_reviews': -1}}
    
query_5 = {'$limit': 10}
    
query_6 = {'$lookup': {
            'from': 'listings', 
            'let': {'city': '$_id'}, 
            'pipeline': [{
                '$match': {'$expr': {'$eq': ['$address.market', '$$city']}} 
            },{
                '$sort': {'number_of_reviews': -1}
            },{
                '$limit': 3
            },{
                '$project': {'_id': 1, 'name': 1, 'price': 1, 'number_of_reviews': 1}
            }],
            'as': 'top3_properties'}
    }

query_7 = {'$project': {
            '_id': 1,
            'total_reviews':1,
            'total_properties': 1,
            'property_types': 1, 
            'top3_properties': 1}
    }

query_8 = {'$merge': {
            'into': 'top10_cities',
            'whenMatched': 'replace'}
    }


pipeline = [query_1, query_2, query_3, query_4, query_5, query_6, query_7, query_8]

result = list(db.listings.aggregate(pipeline))

pprint(db.top10_cities.find_one())
```
